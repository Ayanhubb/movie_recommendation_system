```python
from nltk.stem.porter import PorterStemmer ## stemming tecnique
ps = PorterStemmer()  
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features= 5000 , stop_words="english") ## stop word like is,of,a ,an this will be removed 
## max_features means we need the most frequently used 5k words
```

```python
## now we need to measure thr distance in vectors.. we will not use euclidean distance cause when datas means vectors are more then its fails 
# we will use cosine distance or similarity cause this measures the angels between vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity 
```

```python
cosine_similarity(vectors).shape ## 4806 vectors distance with 4806 vectors
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[1] ## inedx no 1 movies is fetched for every movies similar distance with 4806
```

```python
## how to find the index for the movies 
new_df[new_df["title"] == "Batman Begins"].index[0]
```

```python
## we are now introduce the enumerate function cause there are 4806 no of indexes and if we sc for 110 index or something else 
#then the order will be unsorted 
sorted(list(enumerate(similarity[0])),reverse = True , key = lambda x:x[1])[1:6]
```

```python
def recommend (movie):
    movie_index = new_df[new_df["title"] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True , key = lambda x:x[1])[1:6]
    for i in movies_list:
        print(new_df.iloc[i[0]].title) ## this will show the nearsest 5 movies title noy return thr index
    
```

```python
recommend("Batman Begins")
```

```python
recommend("Avatar")
```
